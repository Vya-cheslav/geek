#### 1. Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
- Найдите их сумму. (на листочке)
- Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)


```python
import math
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def get_sum_v(x):
    z = 0
    for i in x:
        z = z + i**2

    print(math.sqrt(z))  
```

```python
x = [3,0,0]
get_sum_v(x)
```

#### 3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python

p = 5
x1 = []
y1 = []
y2 = []
for i in range(1000):
    _x = i
    x1.append(_x)
    y1.append(-math.sqrt(2*p*_x))
    y2.append(math.sqrt(2*p*_x))
    
plt.plot(x1, y1)
plt.plot(x1, y2)     
```

```python
# график окружности
r = 1

r = r**.5

t = np.linspace (0, 2*np.pi, 1000)
t

x1 = r * np.cos(t)
x2 = r * np.sin(t)

fig, ax = plt.subplots(1)

fig.set_size_inches(6, 6)

ax.plot(x1, x2)

plt.grid()

plt.show()
```

```python
# график эллипса

u=0     #x-position of the center
v=0    #y-position of the center
a=2.     #radius on the x-axis
b=1.5    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)

plt.figure(figsize=(8, 6))

plt.plot( u+a*np.cos(t) , v+b*np.sin(t) )
plt.grid(color='lightgray',linestyle='--')

plt.show()
```

- Нарисуйте трехмерный график двух параллельных плоскостей.

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,.5)
y = np.arange(-5,5,.5)
X, y = np.meshgrid(X, y)
z = 2*X + 5*y
ax.plot_wireframe(X,y,z)
#ax.scatter(0,0,0,'z',50,'red')
X = np.arange(-5,5,.5)
y = np.arange(-5,5,.5)
X, y = np.meshgrid(X, y)
z = 100+(2*X +5*y)
ax.plot_wireframe(X,y,z, color = 'red')
show()
```
