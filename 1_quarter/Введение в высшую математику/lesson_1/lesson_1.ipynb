```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#создадим массив
x = np.linspace(0, 10, 121)
#Создадим график
fig=plt.figure()
#На графике 2 фигуры
ax=fig.add_subplot(111, label="1")
ax2=fig.add_subplot(111, label="1", frame_on=False)
#Заполним эти фигуры
k = 1
ax.plot(x, np.cos(k*x), color="C0")
k = 1.5
ax2.plot(x, np.cos(k*x), color="C1")
plt.show()
```
