1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.


```python
def average(arr):
    n = len(arr)
    if n == 0:
        return 0
    s = 0
    for i in arr:
        s += i
    return s/n      
```

```python
def standard_deviation(arr, bias = False):
    n = len(arr)-1 if bias == True else len(arr)    
    if n == 0:
        return 0
    avr = average(arr)
    s = 0
    for i in arr:
        s += (i-avr)**2
    return (s/n)**0.5  
```

```python
def dispersion_biased(arr):
    return standard_deviation(arr)**2
```

```python
def variance_is_not_biased(arr):
    return standard_deviation(arr, True)**2
```

```python
arr = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
print(f'Среднее арифметическое = {average(arr)}')
print(f'Среднее квадратичное отклонение = {standard_deviation(arr)}')
print(f'Несмещенная оценка дисперсии = {dispersion_biased(arr)}')
print(f'Смещенная оценка дисперсии = {variance_is_not_biased(arr)}')
```

3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
#Веротность попадания одного из трех составляет 1/3, выстрел сделан один
#Верояятность перовго 0.9
#Вероятность второго 0.8
#вероятность третьео 0.6
p = 1/3
P1 = (0.9*p)/(0.9*p+0.8*p+0.6*p)
P2 = (0.8*p)/(0.9*p+0.8*p+0.6*p)
P3 = (0.6*p)/(0.9*p+0.8*p+0.6*p)
print(f'Вероятность того, что выстрел произведен первым спортсменом = {P1}')
print(f'Вероятность того, что выстрел произведен вторым спортсменом = {P2}')
print(f'Вероятность того, что выстрел произведен третьим спортсменом = {P3}')
```

4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
#Всего на факультете А(25%) + В(25) + С(50%) = 100%
#Вероятность сдачи сессии для А 0.8
#Вероятность сдачи сессии для В 0.7
#Вероятность сдачи сессии для С 0.9
p_A=1/4
p_B=1/4
p_C=1/2
P1 = (0.8*p_A)/(0.8*p_A+0.7*p_B+0.9*p_C)
P2 = (0.7*p_B)/(0.8*p_A+0.7*p_B+0.9*p_C)
P3 = (0.9*p_C)/(0.8*p_A+0.7*p_B+0.9*p_C)
print(f'Вероятность того, что студент учится на факультете А = {P1}')
print(f'Вероятность того, что студент учится на факультете В = {P2}')
print(f'Вероятность того, что студент учится на факультете С = {P3}')
```
